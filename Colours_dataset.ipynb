```
import requests
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import random
import pandas as pd
import numpy as np
import time
import json
import time



def query(color):


  data = {'model':"default",'input':[color,"N","N","N","N"]}
  data = json.dumps(data)

  response = requests.post('http://colormind.io/api/', data=data)
  return response


def msklist(n):
  for _ in range(n):
    yield []

arr = []

dictionaryofColors0 = dict(zip(("red0","green0", "blue0","hex0"),([],[],[],[])))
dictionaryofColors1 = dict(zip(("red1","green1", "blue1","hex1"),([],[],[],[])))
dictionaryofColors2 = dict(zip(("red2","green2", "blue2","hex2"),([],[],[],[])))
dictionaryofColors3 = dict(zip(("red3","green3", "blue3","hex3"),([],[],[],[])))
dictionaryofColors4 = dict(zip(("red4","green4", "blue4","hex4"),([],[],[],[])))


arr.append(dictionaryofColors0)
arr.append(dictionaryofColors1)
arr.append(dictionaryofColors2)
arr.append(dictionaryofColors3)
arr.append(dictionaryofColors4)




data = []
counter =0
first_time =0
df = pd.DataFrame()

def generate_rgb():
  return (random.uniform(0,1),random.uniform(0,1), random.uniform(0,1))



def converto_intrgb(input):
  return (float(input[0]/255), float(input[1]/255),float(input[2]/255))





def convert_to_rgb(input):
  return (int(input[0] * 255), int(input[1] * 255), int(input[2] * 255))

def rgbtohex(r,g,b):
  return "#{:02x}{:02x}{:02x}".format(r,g,b)


def generate_patches(hex):
  colors = []
  for i in range(5):
    colors.append([hex[i]])

  categories = ["color 1", "color 2", "color 3", "color 4", "color 5"]
  legend_dict = dict(zip(categories,colors))
  patchesM = []
  for key in legend_dict:
    data_key = patches.Patch(facecolor=legend_dict[key][0],edgecolor=None,label=key)
    patchesM.append(data_key)

  plt.gca()
  plt.legend(handles=patchesM,ncol=len(categories), fontsize='small')
  plt.show()


```

```
def colorDataCreator(first_time,color,count):


  global counter,arr, df
  
  for i in range(count):
      time.sleep(3)
      hex = []
      result  = query(color).json().get("result")

      print(result)
      print(arr)
      print("Result "+str(i))
      

      for i in range(len(result)):
        hex.append(rgbtohex(result[i][0], result[i][1], result[i][2]))
        arr[i].get("red"+str(i)).append(result[i][0])
        arr[i].get("green" + str(i)).append(result[i][1])
        arr[i].get("blue"+str(i)).append(result[i][2])
        arr[i].get("hex"+str(i)).append(rgbtohex(result[i][0], result[i][1], result[i][2]))
      
      generate_patches(hex)

  
 #indian shit to fix
  dataforDf = {'red0': arr[0].get("red0"), 'green0': arr[0].get("green0"), 'blue0': arr[0].get("blue0"),
               'hex0': arr[0].get("hex0"),
               'red1': arr[1].get("red1"), 'green1': arr[1].get("green1"),'blue1': arr[1].get("blue1"),
               'hex1':arr[1].get("hex1"),
               'red2': arr[2].get("red2"), 'green2': arr[2].get("green2"), 'blue2': arr[2].get("blue2"),
               "hex2":arr[2].get("hex2"),
               'red3': arr[3].get("red3"), 'green3': arr[3].get("green3"),'blue3': arr[3].get("blue3"),
               "hex3": arr[3].get("hex3"),
               'red4': arr[4].get("red4"), 'green4': arr[4].get("green4"), 'blue4': arr[4].get("blue4"),
               "hex4": arr[4].get("hex4")}

  df = pd.DataFrame(data=dataforDf)
  if first_time ==1:

    #df.drop_duplicates(subset=['hex0'], keep='first', inplace=True)
    df.to_csv('data.csv', index=False, header=True)

    
  else:

    df.drop_duplicates(keep="first",inplace=True)
    df.to_csv('data.csv', index=False, mode='a', header=False)
    


```
